```python
from sklearn import datasets
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
import numpy as np
from sklearn.grid_search import GridSearchCV
import math
```

```python
bean = datasets.load_boston()
print bean.DESCR
```

```python
def load_boston():
    scaler = StandardScaler()
    boston = datasets.load_boston()
    X=boston.data
    y=boston.target
    X = scaler.fit_transform(X)
    return train_test_split(X,y)
    
```

```python
X_train, X_test, y_train, y_test = load_boston()
```

```python
X_train.shape
```

### Fitting a Linear Regression

It's as easy as instantiating a new regression object (line 1) and giving your regression object your training data
(line 2) by calling .fit(independent variables, dependent variable)



```python

clf = LinearRegression()
clf.fit(X_train, y_train)
```

### Making a Prediction
X_test is our holdout set of data.  We know the answer (y_test) but the computer does not.   

Using the command below, I create a tuple for each observation, where I'm combining the real value (y_test) with
the value our regressor predicts (clf.predict(X_test))

Use a similiar format to get your r2 and mse metrics working.  Using the [scikit learn api](http://scikit-learn.org/stable/modules/model_evaluation.html) if you need help!

```python
zip (y_test, clf.predict(X_test))
```

R2 Score:

```python
r2_score(y_test, clf.predict(X_test))
```

RMSE:

```python
math.sqrt(mean_squared_error(y_test, clf.predict(X_test)))
```

### Using L2 


```python
cle =Ridge(alpha=1)
cle.fit(X_train, y_train)
```

```python
zip (y_test, cle.predict(X_test))
```

R2 Score:

```python
r2_score(y_test, cle.predict(X_test))
```

RMSE:

```python
math.sqrt(mean_squared_error(y_test, cle.predict(X_test)))
```

###Using GridSearchCV to attempt to find the best value of alpa for ridge

```python
alphaValues = np.array([1.5,1.4,1.2,1.1,1.3,0.99,1,0.98,0.97,0.965,0.954,00.955])
rl2 = Ridge()
bestModel = GridSearchCV(estimator=rl2,  param_grid=dict(alpha=alphaValues))
bestModel.fit(X_train, y_train)
```

```python
zip (y_test, bestModel.predict(X_test))
```

R2 Score:

```python
r2_score(y_test, bestModel.predict(X_test))
```

RMSE:

```python
math.sqrt(mean_squared_error(y_test, bestModel.predict(X_test)))
```
