```python
import pandas as pd
%pylab inline
```

```python
df = pd.read_csv("train.csv")
```

```python
df.columns
```

```python
X = pd.DataFrame()
X['sex'] = df['Sex']
X['age'] = df['Age']
X['sibsp'] = df['SibSp']
X['parch'] = df['Parch']
X['class'] = df ['Pclass']
X['fare'] = df['Fare']
X['Embarked'] = df ['Embarked']
X['survived'] = df['Survived']
X['cabin'] = df['Cabin']


```

```python
def clean_cabin(x):
    try:
        return x[0]
    except TypeError:
        return "None"
X['cabin'] = X.cabin.apply(clean_cabin)
```

```python
X['age'].fillna(X.age.mean(), inplace=True)

X.describe()
```

Use dummy variables to replace categorical variables.

```python
categorical_variables = ['sex', 'cabin', 'Embarked']
for variable in categorical_variables:
    X[variable].fillna("Missing", inplace=True)
    dummies = pd.get_dummies(X[variable], prefix=variable)
    X = pd.concat([X, dummies], axis=1)
    X.drop([variable], axis=1, inplace=True)
   
```

Drop co-linear variable

```python
X.drop(['sex_male'],axis=1)
X.drop(['Embarked_S'],axis=1)
X.drop(['cabin_None'],axis=1)
```
