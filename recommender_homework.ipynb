```python
def pearson(s1, s2):
    """Take two pd.Series objects and return a pearson correlation."""
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))
```

```python
def get_recs(movie_name, M, num):

    import numpy as np
    reviews = []
    for title in M.columns:
        if title == movie_name:
            continue
        cor = pearson(M[movie_name], M[title])
        if np.isnan(cor):
            continue
        else:
            reviews.append((title, cor))
    
    reviews.sort(key=lambda tup: tup[1], reverse=True)
    return reviews[:num]

    
```

The following is a function that takes a data frame and a movie name and returns the average rating. I will pass in the data frame with people who appear to have movie tastes similar to mine (similarTaste)


```python
def get_prediction(movie_name, df):
    import numpy as np
    import scipy
    
    rating = []
    for id, value in df.iterrows():
        if value.Movie == movie_name:            
            rating.append(value.Rating)
    return scipy.mean(rating)
```
